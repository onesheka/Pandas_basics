### Making changes to the data

```python
## Let's look at the dataset
df.head(5)
## now one thing we can do to the dataset is add all the int values from HP to Speed and put it in a new column total which would later help us find which Pokemon is the best say
```

```python
df['Total'] = df.iloc[:,4:10].sum(axis = 1)
print(df.head(5))
```

```python
cols = list(df.columns) ##to rearrange the columns to make it more readable
df = df[cols[0:4] + [cols[-1]] + cols[4:12]] ## when data is changing, don't use the hardcoded numbers
print(df.head(5))
```

```python
df.to_csv('modified_csv',index=False) ## turning the dataset back to the csv files
```

### Filtering Data

```python
new_df = df.loc[(df['Type 1']=='Grass') & (df['Type 2']=='Poison') & (df['HP']>70)]
new_df.to_csv('new_modified_csv',index=False)
#print(new_df.head(5))
new_df = new_df.reset_index() ### to reset the index values
new_df
```

### Now, if we want to choose all the names that contain the word Mega

```python
df.loc[~df['Name'].str.contains('Mega')] ##a;; names that do not  contain Mega Keyword(~ sumbol)
```

### using regular impressions

```python
import re

df.loc[df['Type 1'].str.contains('Fire|Grass', regex = True)] ##using the regular expressions
```

```python
#### If we don't remember the capital letters
import re
df.loc[df['Type 1'].str.contains('Fire|Grass',flags = re.I, regex = True)] ## ignoring the cases
```

### Aggregate Statistics (Groupby)

```python
df.groupby('Type 1').mean().sort_values('Defense',ascending=False) ## to group all the types of pokemon by their mean
## mean, sum, count
```

### Working with large amount of dataset

```python
for df in pd.read_csv('modified_csv',chunksize=5): ##using chunksize, we can work with chunks to break down the rows and data we are working with
    print("DATAFRAME CHUNK")
    print(df)
```
